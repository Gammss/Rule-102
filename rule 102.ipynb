```python
# Importing the required modules
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
 
plt.rcParams["figure.figsize"]=(7,5)
# Generating data for the heat map
data = np.random.rand( 10 , 10 )

# Function to show the heat map
ax = sns.heatmap( data , cmap = 'magma' )
  
# Adding details to the plot
plt.title( "2-D Heat Map" )
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# Displaying the plot
plt.show()
```

```python

rows = 100
columns = 99

midpoint = int((columns + 1)/2)

grid = np.zeros((rows, columns), dtype=int)
grid[0, 50] = 1
heat = sns.heatmap(grid, cmap = 'magma')
```

```python
x = 1
y = 0
while x < rows:
    y = 0
    while y < columns:
        # print (str(y) + ", " + str (x))
        if (y < columns):

            if y > 0:
                left = grid[x - 1, y - 1]
            elif y == 0:
                left = grid[x - 1, columns - 1]
        
            if y < columns - 1:
                right = grid[x - 1, y + 1]
            elif y == columns - 1:
                right = grid[x - 1, 0]
        
            mid = grid[x - 1, y]

            # print(str(left) + " " + str(mid) + " " + str(right))

        if left == 1:
            if mid == 1:
                if right == 1:
                    grid[x, y] = 0
                elif right == 0:
                    grid[x, y] = 1
            elif mid == 0:
                if right == 1:
                    grid[x, y] = 1
                elif right == 0:
                    grid[x, y] = 0
        elif left == 0:
            if mid == 1:
                if right == 1:
                    grid[x, y] = 0
                elif right == 0:
                    grid[x, y] = 1
            elif mid == 0:
                if right == 1:
                    grid[x, y] = 1
                elif right == 0:
                    grid[x, y] = 0
        y+= 1
    x += 1
heat = sns.heatmap(grid, cmap = 'magma')
```
